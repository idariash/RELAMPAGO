```python
import pyart
import time

def grid_radar(radar, grid_shape=(20, 401, 401), xlim=(-100000, 100000),
               ylim=(-100000, 100000), zlim=(1000, 20000),
               fields=['reflectivity', 'differential_reflectivity'], origin=None):
    radar_list = [radar]
    
    if origin is None:
        origin = (radar.latitude['data'][0],
                  radar.longitude['data'][0])
        
    grid = pyart.map.grid_from_radars(
        radar_list, grid_shape=grid_shape,
        grid_limits=(zlim, ylim, xlim),
        grid_origin=origin, fields=fields,
        gridding_algo='map_gates_to_grid', grid_origin_alt=0.0)

    
    return grid

'/net/denali/storage/radar/RELAMPAGO/quality_controlled_data/level_1a/2018/12/14/chivo.1a.20181214_010047.REL_PFAR360.nc'
chivo_filename = '/net/denali/storage/radar/RELAMPAGO/quality_controlled_data/level_1a/2018/12/14/chivo.1a.20181214_013054.REL_PFAR360.nc'
'/net/denali/storage/radar/RELAMPAGO/quality_controlled_data/level_1a/2018/12/14/chivo.1a.20181214_014059.REL_PFAR360.nc'
'/net/denali/storage/radar/RELAMPAGO/quality_controlled_data/level_1a/2018/12/14/chivo.1a.20181214_020047.REL_PFAR360.nc'

'/net/denali/storage/radar/CSAPR/DOE_a1/corcsapr2cfrppiM1.a1.20181214.010004.nc'
csapr_filename = '/net/denali/storage/radar/CSAPR/DOE_a1/corcsapr2cfrppiM1.a1.20181214.013003.nc'
'/net/denali/storage/radar/CSAPR/DOE_a1/corcsapr2cfrppiM1.a1.20181214.014504.nc'
'/net/denali/storage/radar/CSAPR/DOE_a1/corcsapr2cfrppiM1.a1.20181214.020004.nc'

chivo = pyart.io.read(chivo_filename)
csapr = pyart.io.read(csapr_filename)

chivo_grid = grid_radar(chivo, origin=(chivo.latitude['data'][0], chivo.longitude['data'][0]))
csapr_grid = grid_radar(csapr, origin=(chivo.latitude['data'][0], chivo.longitude['data'][0]))
```

```python
chivo_grid_fname = 'chivo_grid.nc'
csapr_grid_fname = 'csapr_grid.nc'

chivo_grid.write(chivo_grid_fname)
csapr_grid.write(csapr_grid_fname)
```

```python
import scipy.ndimage as spyi
import matplotlib.pyplot as plt
import numpy as np

ilev = 4

fig = plt.figure(figsize=(6, 5))
x, y = np.meshgrid(0.001*chivo_grid.x['data'], 0.001*chivo_grid.y['data'])
Z = chivo_grid.fields['reflectivity']['data']

# Smooth out and mask
#DZcomp = spyi.gaussian_filter(DZcomp, sigma = 2)


Z = np.ma.masked_where( Z < 10, Z)

cs = plt.pcolormesh(x, y, Z[ilev], vmin=10, vmax=60, cmap='pyart_NWSRef')

fig = plt.figure(figsize=(6, 5))
x, y = np.meshgrid(0.001*csapr_grid.x['data'], 0.001*csapr_grid.y['data'])
Z = csapr_grid.fields['reflectivity']['data']

# Smooth out and mask
#DZcomp = spyi.gaussian_filter(DZcomp, sigma = 2)


Z = np.ma.masked_where( Z < 10, Z)

cs = plt.pcolormesh(x, y, Z[ilev], vmin=10, vmax=60, cmap='pyart_NWSRef')
```
